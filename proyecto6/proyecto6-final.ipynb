````python
df.dropna(subset=['name', 'genre'], inplace=True)
print(df.isnull().sum())
````

'''con la columna 'year_of_release' pasamos a convertirla a datetime para poder manipular y estudiar mejor '''

````python
media_lanzamiento = df['year_of_release'].mean()
print(f"La media de la columna 'year_of_release' es: {media_lanzamiento}")
````

'''para poder pasarlo a entero no puede haber valores nulos, como en el ejercicio anterior vemos que hay 269 valores nulos, no son muchos pero tampoco son pocos por lo que decidimos aplicar un .mean() para sacar el valor de la media'''

````python
df['year_of_release'] = df['year_of_release'].fillna(2006)
df['year_of_release'] = df['year_of_release'].astype(int)
df['year_of_release'] = pd.to_datetime(df['year_of_release'].astype(str))

df.info()
````

````python
score_median = df['critic_score'].median()
print(score_median)
````

````python
df['critic_score'] = df['critic_score'].fillna(71.0)
df.info()
````

````python
print(df['user_score'].sample(5))
````

````python
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
#pasamos a comprobar la mediana para llenar los numeros nulos
user_median = df['user_score'].median()
print(user_median)
````

````python
#llenamos la los valores nulos
df['user_score'] = df['user_score'].fillna(7.5)
df.info()
````

````python
#realizaremos el mismo proceso para la columna faltante
print(df['rating'].sample(15))
````

'''decidimos no dejar valores nulos, hemos cambiado dependiendo de cada columna ya sea el datetime, a float64, llenado los valores ausentes dependiendo de cada categoria para sacar el maximo provecho a la informacion, en algunos caso casos como el rating creemos que hay muchos valores nan ya se por que se lleno el formato mal o cumplieron con ciertos requisistos para entrar la informacion de la clasificacion como el mayor numero era lo de los valores nulos decidimos llenarlo por valores unkown y no alterar la informacion que si esta bien'''

<div class="alert alert-block alert-danger">
<b>Comentario de Revisor            </b> <a class="tocSkip"></a>

Por lo general, no es buena idea rellenar todos los ausentes por la media o mediana de toda la data. Lo ideal es realizar una segmentación de los datos nulos y rellenar los nulos según al segmento que pertenece. En el caso de `user_score` y `critic_score`, si revisas la distribución antes y después de imputar, verás que se introdujo un sesgo muy fuerte en la data, la distribución queda marcadamente sobrerepresentada en el valor de la media. 

No tenemos suficiente información para imputar las variables de score, sin embargo, su porcentaje es muy grande y perderíamos la información adyacente si decidieramos eliminar esas filas. Lo mejor es no eliminar los datos ausentes y conservar las columnas tal y como están. O hacer como en el caso de rating, donde las marcaste con unknown.

Por otro lado, para year_of_release se podría segmentar por consola y luego rellenar los datos según la mediana de los años de lanzamiento de la consola en cuestión. Por ejemplo, la mediana de toda la data podría ser 2005, pero si hay un nulo de PS4, no hace sentido rellenarlo por ese año, ya que la consola ni siquiera existía. En ese caso, la mediana de los años de los juegos de PS4 sería mejor candidato de imputación.

````python
df_wii = df[(df['platform'] == 'Wii') &
            (df['critic_score'].notna()) &
            (df['user_score'].notna())]
sns.set_style('whitegrid')
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1) 
sns.scatterplot(x='critic_score', y='total_sales', data=df_wii, alpha=0.6)
plt.title('Críticos vs. Ventas en Wii', fontsize=14)
plt.xlabel('Puntuación de los Críticos (de 0 a 100)', fontsize=11)
plt.ylabel('Ventas Totales (en millones)', fontsize=11)
plt.subplot(1, 2, 2) 
sns.scatterplot(x='user_score', y='total_sales', data=df_wii, alpha=0.6)
plt.title('Usuarios vs. Ventas en Wii', fontsize=14)
plt.xlabel('Puntuación de los Usuarios (de 0 a 10)', fontsize=11)
plt.ylabel('Ventas Totales (en millones)', fontsize=11)

plt.tight_layout()
plt.show()
correlacion_criticos = df_wii['critic_score'].corr(df_wii['total_sales'])
correlacion_usuarios = df_wii['user_score'].corr(df_wii['total_sales'])

print(f'Correlación entre la Puntuación de Críticos y las Ventas: {correlacion_criticos:.2f}')
print(f'Correlación entre la Puntuación de Usuarios y las Ventas: {correlacion_usuarios:.2f}')
````

# Hipótesis
### Hipótesis Nula: Esta es la suposición inicial que queremos desafiar. Afirma que no hay diferencia significativa entre los dos grupos que estamos comparando. Por ejemplo: "La calificación promedio de los usuarios para Xbox One y PC es la misma".
### Hipótesis Alternativa : Es lo que queremos probar. Afirma que sí existe una diferencia significativa. Por ejemplo: "La calificación promedio de los usuarios para Xbox One y PC es diferente".

# El Criterio
Para probar las hipótesis, me fijé en un valor llamado p-valor. Imagina que el p-valor es la probabilidad de que la diferencia que vimos en los datos haya sido pura coincidencia.
Si el p-valor es muy bajo (menor a 0.05), es como si te dijera que es casi imposible que esa diferencia haya sido por suerte. Por lo tanto, decimos: "Esta diferencia es real, no es casualidad".
Si el p-valor es alto (mayor a 0.05), significa que es bastante probable que la diferencia que vimos haya sido solo cuestión de suerte o de azar. Entonces, la conclusión es: "No podemos decir con seguridad que haya una diferencia real".
